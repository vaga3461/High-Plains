```python
# this notebook will test whether channels both widen and steepen under high supply conditions
```

```python
# start by plotting Qs for four sets of runs; 
# this will determine which parameters groups are experiencing increased sediment discharge under varying conditions
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dx = 500
x = np.arange(0, 50000, dx)
area = (1./3.) * x**2
```

```python
# import sediment transport
Qs_default = np.genfromtxt('Qs csv files/Qs_default.csv', delimiter=',')
Qs_default_softsed = np.genfromtxt('Qs csv files/Qs_default_softsed.csv', delimiter=',')
# Qs_1 = np.genfromtxt('Qs csv files/Qs_exp_1.csv', delimiter=',')
Qs_2 = np.genfromtxt('Qs csv files/Qs_exp_2.csv', delimiter=',')
Qs_3 = np.genfromtxt('Qs csv files/Qs_exp_3.csv', delimiter=',')
Qs_4 = np.genfromtxt('Qs csv files/Qs_exp_4.csv', delimiter=',')
Qs_5 = np.genfromtxt('Qs csv files/Qs_exp_5.csv', delimiter=',')
Qs_6 = np.genfromtxt('Qs csv files/Qs_exp_6.csv', delimiter=',')
Qs_7 = np.genfromtxt('Qs csv files/Qs_exp_7.csv', delimiter=',')
Qs_8 = np.genfromtxt('Qs csv files/Qs_exp_8.csv', delimiter=',')
Qs_9 = np.genfromtxt('Qs csv files/Qs_exp_9.csv', delimiter=',')
Qs_10 = np.genfromtxt('Qs csv files/Qs_exp_10.csv', delimiter=',')
Qs_11 = np.genfromtxt('Qs csv files/Qs_exp_11.csv', delimiter=',')
Qs_12 = np.genfromtxt('Qs csv files/Qs_exp_12.csv', delimiter=',')
Qs_13 = np.genfromtxt('Qs csv files/Qs_exp_13.csv', delimiter=',')
Qs_14 = np.genfromtxt('Qs csv files/Qs_exp_14.csv', delimiter=',')
Qs_15 = np.genfromtxt('Qs csv files/Qs_exp_15.csv', delimiter=',')
Qs_16 = np.genfromtxt('Qs csv files/Qs_exp_16.csv', delimiter=',')
```

```python
# import slopes
slope_default = np.genfromtxt('slope csv files/slope_default.csv', delimiter=',')
slope_default_softsed = np.genfromtxt('slope csv files/slope_default_softsed.csv', delimiter=',')
slope_13 = np.genfromtxt('slope csv files/slope_exp_13.csv', delimiter=',')
slope_14 = np.genfromtxt('slope csv files/slope_exp_14.csv', delimiter=',')
slope_15 = np.genfromtxt('slope csv files/slope_exp_15.csv', delimiter=',')
slope_16 = np.genfromtxt('slope csv files/slope_exp_16.csv', delimiter=',')
```

```python
# and widths
width_default = np.genfromtxt('width csv files/b_default.csv', delimiter=',')
width_default_softsed = np.genfromtxt('width csv files/b_default_softsed.csv', delimiter=',')
width_13 = np.genfromtxt('width csv files/b_exp_13.csv', delimiter=',')
width_14 = np.genfromtxt('width csv files/b_exp_14.csv', delimiter=',')
width_15 = np.genfromtxt('width csv files/b_exp_15.csv', delimiter=',')
width_16 = np.genfromtxt('width csv files/b_exp_16.csv', delimiter=',')
```

```python
# now plot them up as experiments

# now make a pretty figure
plt.rcParams['figure.dpi'] = 300
fig, axs = plt.subplots(2, 2, figsize = (14, 10))

fig.text(0.5, 0.05, 'Distance', ha='center', fontsize = 20)
fig.text(0.05, 0.5, 'Sed Supply', va='center', rotation='vertical', fontsize = 20)

axs[0, 0].plot(x, Qs_default, label = 'Default', color = 'goldenrod', linewidth = 3)
axs[0, 0].plot(x, Qs_default_softsed, label = 'Default with soft sed', color = 'red', linewidth = 3)
# axs[0, 0].plot(x, Qs_1, color = 'k', linestyle = '-')
axs[0, 0].plot(x, Qs_2, color = 'k', linestyle = '--')
axs[0, 0].plot(x, Qs_3, color = 'k', linestyle = '-.')
axs[0, 0].plot(x, Qs_4, color = 'k', linestyle = ':')
axs[0, 0].set_title('A', fontsize = 18, x=-0.1,y=1.0)
axs[0, 0].tick_params(axis="x", labelsize=12) 
axs[0, 0].tick_params(axis="y", labelsize=12)
axs[0, 0].legend(fontsize = 12)

axs[0, 1].plot(x, Qs_default, label = 'Default', color = 'goldenrod', linewidth = 3)
axs[0, 1].plot(x, Qs_default_softsed, label = 'Default with soft sed', color = 'red', linewidth = 3)
axs[0, 1].plot(x, Qs_5, color = 'k', linestyle = '-')
axs[0, 1].plot(x, Qs_6, color = 'k', linestyle = '--')
axs[0, 1].plot(x, Qs_7, color = 'k', linestyle = '-.')
axs[0, 1].plot(x, Qs_8, color = 'k', linestyle = ':')
axs[0, 1].set_title('B', fontsize = 18, x=-0.1,y=1.0)
axs[0, 1].tick_params(axis="x", labelsize=12) 
axs[0, 1].tick_params(axis="y", labelsize=12)
axs[0, 1].legend(fontsize = 12)

axs[1, 0].plot(x, Qs_default, label = 'Default', color = 'goldenrod', linewidth = 3)
axs[1, 0].plot(x, Qs_default_softsed, label = 'Default with soft sed', color = 'red', linewidth = 3)
axs[1, 0].plot(x, Qs_9, color = 'k', linestyle = '-')
axs[1, 0].plot(x, Qs_10, color = 'k', linestyle = '--')
axs[1, 0].plot(x, Qs_11, color = 'k', linestyle = '-.')
axs[1, 0].plot(x, Qs_12, color = 'k', linestyle = ':')
axs[1, 0].set_title('C', fontsize = 18, x=-0.1,y=1.0)
axs[1, 0].tick_params(axis="x", labelsize=12) 
axs[1, 0].tick_params(axis="y", labelsize=12)
axs[1, 0].legend(fontsize = 12)

axs[1, 1].plot(x, Qs_default, label = 'Default', color = 'goldenrod', linewidth = 3)
axs[1, 1].plot(x, Qs_default_softsed, label = 'Default with soft sed', color = 'red', linewidth = 3)
axs[1, 1].plot(x, Qs_13, color = 'k', linestyle = '-')
axs[1, 1].plot(x, Qs_14, color = 'k', linestyle = '--')
axs[1, 1].plot(x, Qs_15, color = 'k', linestyle = '-.')
axs[1, 1].plot(x, Qs_16, color = 'k', linestyle = ':')
axs[1, 1].set_title('D', fontsize = 18, x=-0.1,y=1.0)
axs[1, 1].tick_params(axis="x", labelsize=12) 
axs[1, 1].tick_params(axis="y", labelsize=12)
axs[1, 1].legend(fontsize = 12)

#plt.savefig('width fig.png', bbox_inches='tight')
plt.show()
```

```python
# okay, since changing gamma (scenario D) leads to the biggest changes in sed supply, 
# let's see if it also leads to steeper channels

plt.plot(area[1:], slope_default, label = '$\gamma = 0.5$', color = 'goldenrod', linewidth = 3)
plt.plot(area[1:], slope_default_softsed, label = '$\gamma = 0.5$', color = 'red', linewidth = 3)
plt.plot(area[1:], slope_13, label = '$\gamma = 0.25$', color = 'k', linestyle = '-')
plt.plot(area[1:], slope_14, label = '$\gamma = 0.375$', color = 'k', linestyle = '--')
plt.plot(area[1:], slope_15, label = '$\gamma = 0.625$', color = 'k', linestyle = '-.')
plt.plot(area[1:], slope_16, label = '$\gamma = 0.75$', color = 'k', linestyle = ':')
plt.title('D', fontsize = 18, x=-0.1,y=1.0)
plt.tick_params(axis="x", labelsize=12) 
plt.tick_params(axis="y", labelsize=12)
plt.yscale('log')
plt.xscale('log')
plt.legend(fontsize = 12)
```

```python
# interesting, we only get steeper at high supply rates downstream after crossing over "tools/cover" boundary
```

```python
# finally, check out widths
plt.plot(area[1:], width_default, label = '$\gamma = 0.5$', color = 'goldenrod', linewidth = 3)
plt.plot(area[1:], width_default_softsed, label = '$\gamma = 0.5$', color = 'red', linewidth = 3)
plt.plot(area[1:], width_13, label = '$\gamma = 0.25$', color = 'k', linestyle = '-')
plt.plot(area[1:], width_14, label = '$\gamma = 0.375$', color = 'k', linestyle = '--')
plt.plot(area[1:], width_15, label = '$\gamma = 0.625$', color = 'k', linestyle = '-.')
plt.plot(area[1:], width_16, label = '$\gamma = 0.75$', color = 'k', linestyle = ':')
plt.title('D', fontsize = 18, x=-0.1,y=1.0)
plt.tick_params(axis="x", labelsize=12) 
plt.tick_params(axis="y", labelsize=12)
plt.yscale('log')
plt.xscale('log')
plt.legend(fontsize = 12)
```

```python
# okay, now just to make a lil 3 panel figure, like Phillips

plt.rcParams['figure.dpi'] = 300
fig, axs = plt.subplots(2, 2, figsize = (14, 10), sharex = False, sharey = False)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.2)

axs[0, 0].grid(alpha = 0.2)
axs[0, 0].plot(x/500, Qs_default, color = 'goldenrod', linewidth = 3)
axs[0, 0].plot(x/500, Qs_default_softsed, color = 'dodgerblue', linewidth = 3)
axs[0, 0].plot(x/500, Qs_13, color = 'midnightblue', linestyle = '-')
axs[0, 0].plot(x/500, Qs_14, color = 'midnightblue', linestyle = '--')
axs[0, 0].plot(x/500, Qs_15, color = 'midnightblue', linestyle = '-.')
axs[0, 0].plot(x/500, Qs_16, color = 'midnightblue', linestyle = ':')
axs[0, 0].set_title('A', fontsize = 20, x=0.05,y=0.85)
axs[0, 0].set_xlabel('Distance (km)', fontsize = 16)
axs[0, 0].set_ylabel('Sediment transport rate ($m^3/yr$)', fontsize = 16)
axs[0, 0].tick_params(axis="x", labelsize=14) 
axs[0, 0].tick_params(axis="y", labelsize=14)

axs[0, 1].grid(alpha = 0.2)
axs[0, 1].plot(area[1:], slope_default, color = 'goldenrod', linewidth = 3)
axs[0, 1].plot(area[1:], slope_default_softsed, color = 'dodgerblue', linewidth = 3)
axs[0, 1].plot(area[1:], slope_13, color = 'midnightblue', linestyle = '-')
axs[0, 1].plot(area[1:], slope_14, color = 'midnightblue', linestyle = '--')
axs[0, 1].plot(area[1:], slope_15, color = 'midnightblue', linestyle = '-.')
axs[0, 1].plot(area[1:], slope_16, color = 'midnightblue', linestyle = ':')
axs[0, 1].set_title('B', fontsize = 20, x=0.05,y=0.75)
axs[0, 1].tick_params(axis="x", labelsize=14) 
axs[0, 1].tick_params(axis="y", labelsize=14)
axs[0, 1].set_ylabel('Slope', fontsize = 16)
axs[0, 1].set_yscale('log')
axs[0, 1].set_xscale('log')

axs[1,0].set_visible(False) 

axs[1, 1].grid(alpha = 0.2)
axs[1, 1].plot(area[1:], width_13, label = '$\gamma = 0.25$', color = 'midnightblue', linestyle = '-')
axs[1, 1].plot(area[1:], width_14, label = '$\gamma = 0.375$', color = 'midnightblue', linestyle = '--')
axs[1, 1].plot(area[1:], width_default, label = '$\gamma = 0.5,$' r'$\beta = 5$' 'x $10^{-4}$', color = 'goldenrod', linewidth = 3)
axs[1, 1].plot(area[1:], width_default_softsed, label = '$\gamma = 0.5,$' r'$\beta = 1$' 'x $10^{-3}$', color = 'dodgerblue', linewidth = 3)
axs[1, 1].plot(area[1:], width_15, label = '$\gamma = 0.625$', color = 'midnightblue', linestyle = '-.')
axs[1, 1].plot(area[1:], width_16, label = '$\gamma = 0.75$', color = 'midnightblue', linestyle = ':')
axs[1, 1].set_title('C', fontsize = 20, x=0.05,y=0.85)
axs[1, 1].tick_params(axis="x", labelsize=14) 
axs[1, 1].tick_params(axis="y", labelsize=14)
axs[1, 1].set_yscale('log')
axs[1, 1].set_xscale('log')
axs[1, 1].set_xlabel('Area ($m^2$)', fontsize = 16)
axs[1, 1].set_ylabel('Width (m)', fontsize = 16)
axs[1, 1].legend(fontsize = 14, bbox_to_anchor=(-.40, .80))


# axs[1,1].set_visible(False) 

plt.savefig('channel width 2.pdf', bbox_inches='tight')
plt.show()
```
