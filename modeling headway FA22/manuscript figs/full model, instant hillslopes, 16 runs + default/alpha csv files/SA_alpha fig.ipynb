```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import elevation csvs
default = np.genfromtxt('alpha_default.csv', delimiter=',')
exp_1 = np.genfromtxt('alpha_exp_1.csv', delimiter=',')
exp_2 = np.genfromtxt('alpha_exp_2.csv', delimiter=',')
exp_3 = np.genfromtxt('alpha_exp_3.csv', delimiter=',')
exp_4 = np.genfromtxt('alpha_exp_4.csv', delimiter=',')
exp_5 = np.genfromtxt('alpha_exp_5.csv', delimiter=',')
exp_6 = np.genfromtxt('alpha_exp_6.csv', delimiter=',')
exp_7 = np.genfromtxt('alpha_exp_7.csv', delimiter=',')
exp_8 = np.genfromtxt('alpha_exp_8.csv', delimiter=',')
exp_9 = np.genfromtxt('alpha_exp_9.csv', delimiter=',')
exp_10 = np.genfromtxt('alpha_exp_10.csv', delimiter=',')
exp_11 = np.genfromtxt('alpha_exp_11.csv', delimiter=',')
exp_12 = np.genfromtxt('alpha_exp_12.csv', delimiter=',')
exp_13 = np.genfromtxt('alpha_exp_13.csv', delimiter=',')
exp_14 = np.genfromtxt('alpha_exp_14.csv', delimiter=',')
exp_15 = np.genfromtxt('alpha_exp_15.csv', delimiter=',')
exp_16 = np.genfromtxt('alpha_exp_16.csv', delimiter=',')
```

```python
plt.rcParams['figure.dpi'] = 300
fig, axs = plt.subplots(2, 2, figsize = (14, 10))

fig.text(0.5, 0.05, 'Distance (km)', ha='center', fontsize = 20)
fig.text(0.05, 0.5, 'Transport Capacity', va='center', rotation='vertical', fontsize = 20)

axs[0, 0].plot(1 - default[:-1], label = '$K = 1 x 10 ^{-7}$', color = 'goldenrod', linewidth = 3)
axs[0, 0].plot(1 - exp_1[:-1], label = '$K = 1 x 10 ^{-8}$', color = 'k', linestyle = '-')
axs[0, 0].plot(1 - exp_2[:-1], label = '$K = 5 x 10 ^{-8}$', color = 'k', linestyle = '--')
axs[0, 0].plot(1 - exp_3[:-1], label = '$K = 5 x 10 ^{-7}$', color = 'k', linestyle = '-.')
axs[0, 0].plot(1 - exp_4[:-1], label = '$K = 6 x 10 ^{-7}$', color = 'k', linestyle = ':')
axs[0, 0].set_title('A', fontsize = 18, x=-0.1,y=1.0)
axs[0, 0].tick_params(axis="x", labelsize=12) 
axs[0, 0].tick_params(axis="y", labelsize=12)
axs[0, 0].set_ylim(0, 1)
axs[0, 0].legend(fontsize = 12)

axs[0, 1].plot(1 - default[:-1], label = '$r = 5$', color = 'goldenrod', linewidth = 3)
axs[0, 1].plot(1 - exp_5[:-1], label = '$r = 1$', color = 'k', linestyle = '-')
axs[0, 1].plot(1 - exp_6[:-1], label = '$r = 2.5$', color = 'k', linestyle = '--')
axs[0, 1].plot(1 - exp_7[:-1], label = '$r = 7.5$', color = 'k', linestyle = '-.')
axs[0, 1].plot(1 - exp_8[:-1], label = '$r = 10$', color = 'k', linestyle = ':')
axs[0, 1].set_title('B', fontsize = 18, x=-0.1,y=1.0)
axs[0, 1].tick_params(axis="x", labelsize=12) 
axs[0, 1].tick_params(axis="y", labelsize=12)
axs[0, 1].set_ylim(0, 1)
axs[0, 1].legend(fontsize = 12)

axs[1, 0].plot(1 - default[:-1], label = '$k_{qs} = 0.041$', color = 'goldenrod', linewidth = 3)
axs[1, 0].plot(1 - exp_9[:-1], label = '$k_{qs} = 0.021$', color = 'k', linestyle = '-')
axs[1, 0].plot(1 - exp_10[:-1], label = '$k_{qs} = 0.031$', color = 'k', linestyle = '--')
axs[1, 0].plot(1 - exp_11[:-1], label = '$k_{qs} = 0.051$', color = 'k', linestyle = '-.')
axs[1, 0].plot(1 - exp_12[:-1], label = '$k_{qs} = 0.061$', color = 'k', linestyle = ':')
axs[1, 0].set_title('C', fontsize = 18, x=-0.1,y=1.0)
axs[1, 0].tick_params(axis="x", labelsize=12) 
axs[1, 0].tick_params(axis="y", labelsize=12)
axs[1, 0].set_ylim(0, 1)
axs[1, 0].legend(fontsize = 12)

axs[1, 1].plot(1 - default[:-1], label = '$\gamma = 0.5$', color = 'goldenrod', linewidth = 3)
axs[1, 1].plot(1 - exp_13[:-1], label = '$\gamma = 0.25$', color = 'k', linestyle = '-')
axs[1, 1].plot(1 - exp_14[:-1], label = '$\gamma = 0.375$', color = 'k', linestyle = '--')
axs[1, 1].plot(1 - exp_15[:-1], label = '$\gamma = 0.625$', color = 'k', linestyle = '-.')
axs[1, 1].plot(1 - exp_16[:-1], label = '$\gamma = 0.75$', color = 'k', linestyle = ':')
axs[1, 1].set_title('D', fontsize = 18, x=-0.1,y=1.0)
axs[1, 1].tick_params(axis="x", labelsize=12) 
axs[1, 1].tick_params(axis="y", labelsize=12)
axs[1, 1].set_ylim(0, 1)
axs[1, 1].legend(fontsize = 12)
```
