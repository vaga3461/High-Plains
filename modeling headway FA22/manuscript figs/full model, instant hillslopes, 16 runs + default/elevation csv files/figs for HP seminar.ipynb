```python
import numpy as np
```

```python
# import elevs
eta_default = np.genfromtxt('eta_default.csv', delimiter=',')
eta_hard_sed = np.genfromtxt('eta_hard_sed.csv', delimiter=',')
eta_soft_sed = np.genfromtxt('eta_soft_sed.csv', delimiter=',')
```

```python
import matplotlib.pyplot as plt
```

```python
dx = 500
x = np.arange(0, 50000, dx)
```

```python
plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 4))

plt.plot(x/dx, eta_default, label = 'Uniform abrasion coefficients', color = 'goldenrod', linewidth = 3)
plt.plot(x/dx, eta_hard_sed, label = 'Sediment hard relative to bedrock', color = 'k', linewidth = 3, linestyle = ':')
plt.plot(x/dx, eta_soft_sed, label = 'Sediment soft relative to bedrock', color = 'k', linewidth = 3, linestyle = "--")
plt.xlabel('Distance (km)', fontsize = 14)
plt.ylabel('Elevation (m)', fontsize = 14)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

#plt.savefig('sed_hardness.png', bbox_inches='tight')
```

```python
plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 4))

plt.plot(x[5:60]/dx, eta_default[5:60], label = 'Uniform abrasion coefficients', color = 'goldenrod', linewidth = 3)
plt.plot(x[5:60]/dx, eta_hard_sed[5:60], label = 'Sediment hard relative to bedrock', color = 'k', linewidth = 3, linestyle = ':')
plt.plot(x[5:60]/dx, eta_soft_sed[5:60], label = 'Sediment soft relative to bedrock', color = 'k', linewidth = 3, linestyle = "--")
plt.xlabel('Distance (km)', fontsize = 14)
plt.ylabel('Elevation (m)', fontsize = 14)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.savefig('sed_hardness_profile zoom.png', bbox_inches='tight')
```
