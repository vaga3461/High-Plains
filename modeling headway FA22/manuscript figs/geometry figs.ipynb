```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# for this notebook, just use previous runs to get at width, depth of rivers across scales
```

```python
# start by importing phillips data (already filtered in Excel)
global_data = pd.read_excel('phillips_data_clean.xlsx')
```

```python
global_data
```

```python
# some values
R = 2.65
g = 9.81

# add some columns
global_data['Nondim W'] = global_data['Width (m)']/global_data['D50 (m)']
global_data['Nondim D'] = global_data['Depth (m)']/global_data['D50 (m)']
global_data['Nondim Q'] = global_data['Discharge (m3/s)']/(np.sqrt(R * g * (global_data['D50 (m)']**5)))

# does R belong in dimensionless Q calculation?
# unclear from Phillips et al 2022 (R is dimensionless, so does not effect dimensions of calculation)
```

```python
global_data
```

```python
# great, now make an excel workbook to hold values from experiments
# (this means values for slope, width, and depth for all runs from figs 1 and 2)
# in addition, do a range of Q values by changing r from 1 - 10 (1, 5, and 10)
```

```python
# now add in values from alluvial geometry experiments excel workbook
experiments = pd.read_excel('alluvial_geom_experiments.xlsx')

# and filter out experiments that haven't been completed yet
complete_experiments = experiments.dropna()
```

```python
# great, now put these values on some figures

fig, axs = plt.subplots(2, 2)

axs[0, 0].scatter(global_data['Nondim Q'], global_data['Nondim W'])
axs[0, 0].scatter(complete_experiments['Nondim Q'], complete_experiments['Nondim W'])
axs[0, 0].set_xscale('log')
axs[0, 0].set_yscale('log')

axs[0, 1].scatter(global_data['Nondim Q'], global_data['Nondim D'])
axs[0, 1].scatter(complete_experiments['Nondim Q'], complete_experiments['Nondim D'])
axs[0, 1].set_xscale('log')
axs[0, 1].set_yscale('log')

axs[1, 0].scatter(global_data['Nondim Q'], global_data['Slope'])
axs[1, 0].scatter(complete_experiments['Nondim Q'], complete_experiments['Slope'])
axs[1, 0].set_xscale('log')
axs[1, 0].set_yscale('log')

axs[1,1].set_visible(False) 

plt.show()
```
