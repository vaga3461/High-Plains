```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Hstar = 0.2

H = np.linspace(0, 1, 50)
```

```python
fxn = np.exp(-H/Hstar)
opp_fxn = 1 - fxn

ratio = H/Hstar
```

```python
plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 4))

plt.plot(ratio, opp_fxn, label = 'Sediment transport', color = 'k', linewidth = 3, linestyle = ':')
plt.plot(ratio, fxn, label = 'Bedrock erosion', color = 'k', linewidth = 3, linestyle = '--')
plt.xlabel('Sediment thickness ratio (H/H*)', fontsize = 16)
plt.ylabel('Process efficiency', fontsize = 16)
plt.legend(fontsize = 14)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.annotate(r'$1 - \alpha$', (4.5, 0.9), fontsize = 16, color = 'k')
plt.annotate(r'$\alpha$', (4.6, 0.05), fontsize = 16, color = 'k')

plt.savefig('bed_cover 2.pdf', transparent = True, bbox_inches="tight")
```
