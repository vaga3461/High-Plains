```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Hstar = 0.2

H = np.linspace(0, 1, 50)
```

```python
fxn = np.exp(-H/Hstar)
opp_fxn = 1 - fxn

ratio = H/Hstar
```

```python
plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 4))

plt.plot(ratio, fxn, label = 'Bedrock erosion', color = 'goldenrod', linewidth = 3)
plt.plot(ratio, opp_fxn, label = 'Sediment transport', color = 'midnightblue', linewidth = 3)
plt.xlabel('Sediment thickness ratio (H/H*)', fontsize = 14)
plt.ylabel('Process efficiency', fontsize = 14)
plt.legend(fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

plt.savefig('bed_cover.png', transparent = True)
```
