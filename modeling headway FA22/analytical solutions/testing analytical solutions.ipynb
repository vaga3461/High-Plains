```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 100, 10)

U = 0.001
K = 0.0001
A = 0.3 * x**2
m = 0.5
n = 1.0

slope = (U/(K * A**(m/n)))
```

```python
plt.plot(x, slope)
```

```python
plt.plot(A, slope)
plt.xscale('log')
plt.yscale('log')
```

```python
# okay, works
```

```python
# now let's try the plucking + abrasion solution
```

```python
beta = 0.0004
c = 0.041
runoff = 10.
q = runoff * A

slope_2 = (U / (K + (beta*c))) * (1/q)
```

```python
plt.plot(x, slope_2)
```

```python
plt.plot(A, slope_2)
plt.xscale('log')
plt.yscale('log')
```

```python
# okay, cool
```

```python
# now the big test - is there a solution for the sediment layer?
```

```python
K_array = np.zeros(len(x))
K_array[0:25] = 0.0001
K_array[25:] = 0.001
```

```python
num = (c * x) + (beta * c * x) - (K_array * x)
den = c + (beta * c) - K_array

slope_3 = -0.5 * (num/den)
```

```python
plt.plot(x, slope_3)
```

```python
num = c * q
den_1 = K_array * q * x
den_2 = c
den_3 = beta * c * q * x

slope_3 = num / (den_1 - den_2 - den_3)
```

```python
plt.plot(x, slope_3)
```

```python
plt.plot(A[1:], slope_3[1:])
plt.xscale('log')
plt.yscale('log')
```
