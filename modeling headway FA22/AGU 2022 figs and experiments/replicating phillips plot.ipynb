```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
global_data = pd.read_csv('comparison with phillips_depth.csv')
```

```python
global_data['Dimensionless width'] = global_data['Width (m)'] / global_data['D50 (m)']
global_data['Dimensionless discharge'] = global_data['Discharge (m3/s)']/(np.sqrt(9.81 * (global_data['D50 (m)']**5)))
global_data['Dimensionless depth'] = global_data['Depth (m)'] / global_data['D50 (m)']
```

```python
my_Q_data = [3.26700000e+10]
my_width_data = [54.76017592]


my_Q_nondim = my_Q_data / (np.sqrt(2.65 * 9.81 * (0.015)**5))
my_width_nondim = np.array(my_width_data) / 0.015
```

```python
plt.rcParams['figure.dpi'] = 100
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 5))

plt.scatter(global_data['Dimensionless discharge'], global_data['Dimensionless width'], marker = 'x')
plt.scatter(my_Q_nondim, my_width_nondim, marker = 'x', color = 'red')

plt.xlim(1e1, 1e15)
plt.ylim(1e1, 1e7)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Dimensionless discharge $Q/\sqrt{(g D_{50}^5)}$', fontsize = 14)
plt.ylabel('Dimensionless width $W/D_{50}$', fontsize = 14)
plt.show()
```
