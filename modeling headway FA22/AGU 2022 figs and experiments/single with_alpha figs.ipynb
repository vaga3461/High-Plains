```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RUN 1: 
# K = 1e-7
# beta = 1e-7
# psi = 1e-7
# gamma = 0.3
# r = 10.
# D = 0.03
# uplift = 0.0001

alpha_run_1 = np.genfromtxt('alpha run 1.csv')
S_run_1 = np.genfromtxt('slope run 1.csv')
eta_run_1 = np.genfromtxt('topo run 1.csv')
etab_run_1 = np.genfromtxt('bedrock run 1.csv')
```

```python
# RUN 2: 
# K = 1e-8
# beta = 1e-7
# psi = 1e-7
# gamma = 0.3
# r = 10.
# D = 0.03
# uplift = 0.0001

alpha_run_2 = np.genfromtxt('alpha run 2.csv')
S_run_2 = np.genfromtxt('slope run 2.csv')
eta_run_2 = np.genfromtxt('topo run 2.csv')
etab_run_2 = np.genfromtxt('bedrock run 2.csv')
```

```python
plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300

plt.figure(figsize = (8, 4))

plt.plot(1 - alpha_run_1[:-1], label = 'Weaker rock (more pluckable; K = 1E-7)', color = 'midnightblue', linewidth = 3)
plt.plot(1 - alpha_run_2[:-1], label = 'Stronger rock (less pluckable; K = 1E-8)', color = 'silver', linewidth = 3)
plt.xlabel('Distance (km)', fontsize = 16)
plt.ylabel('Sediment Transport Efficiency', fontsize = 16)
plt.hlines(1, xmin = 0, xmax = 100, linestyle = ':', color = 'k', label = 'Transport-limited')
plt.legend(fontsize = 14)

plt.savefig('hard rock_soft rock TPL.png')
plt.show()

# plt.plot(etab_run_1[:-1], color = 'blue', label = 'more erodible bedrock (K = 1e-7)')
# plt.plot(eta_run_1[:-1], color = 'blue')
# plt.plot(etab_run_2[:-1], color = 'green', label = 'less erodible bedrock (K = 1e-8)')
# plt.plot(eta_run_2[:-1], color = 'green')
# plt.xlabel('Distance (km)')
# plt.ylabel('Elevation (m)')
# plt.legend()
# plt.show()
```
