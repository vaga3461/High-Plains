##### analytical solution to stream power

$$U = E$$
$$U = KQS$$
$$Q = \text{area} * \text{runoff}$$

$$S = \frac{U}{KQ}$$

```python
r = 10.

Q = area * r

U = 0.001
K = np.zeros(len(x))
K[:24] = 1e-7
K[24:] = 1e-6
```

```python
S_pred = U/(K[1:] * Q[1:])
```
