```python
# quick notebook to test whether single abrasion class case, and 3 class case with uniform coeffs, give same result
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import files
S_uni = np.loadtxt('slope_uni.csv')
S_def = np.loadtxt('slope_def.csv')
S_var = np.loadtxt('slope_var.csv')
S_neg = np.loadtxt('slope_neg.csv')

# import files
Qs_uni = np.loadtxt('Qs_uni.csv')
Qs_def = np.loadtxt('Qs_def.csv')
Qs_var = np.loadtxt('Qs_var.csv')
Qs_neg = np.loadtxt('Qs_neg.csv')
```

```python
# calc area
dx = 500
x = np.arange(0, 50000, dx)
area = (1./3.) * (x**2)
```

```python
# make slope-area plot

plt.figure(figsize = (8, 4), dpi = 300)

plt.plot(area[1:], S_uni, label = 'uniform abrasion')
plt.plot(area[1:], S_var, label = 'variable abrasion')
plt.plot(area[1:], S_neg, label = 'insignificant abrasion')
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.ylabel('log(Slope)')
plt.xlabel('log(Area)')
```

```python
plt.figure(figsize = (8, 4), dpi = 300)

plt.plot(area, Qs_uni, label = 'uniform abrasion')
plt.plot(area, Qs_var, label = 'variable abrasion')
plt.plot(area, Qs_neg, label = 'insignificant abrasion')
plt.legend()
plt.yscale('log')
plt.xscale('log')
plt.ylabel('log(Bedload transport rate)')
plt.xlabel('log(Area)')
```
