```python
# this notebook creates some plots to compare the effects of uniform vs variable abrasion on channel profiles
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import files
S_uni = np.loadtxt('uniform abrasion/slope_uni.csv')
S_var = np.loadtxt('variable abrasion/slope_var.csv')

eta_uni = np.loadtxt('uniform abrasion/eta_uni.csv')
eta_var = np.loadtxt('variable abrasion/eta_var.csv')
```

```python
# calc area
dx = 500
x = np.arange(0, 50000, dx)
area = (1./3.) * (x**2)
```

```python
# make slope-area plot

plt.plot(area[1:], S_uni, label = 'uniform abrasion')
plt.plot(area[1:], S_var, label = 'variable abrasion')
plt.legend()
plt.yscale('log')
plt.xscale('log')
```

```python
plt.plot(x, eta_uni, label = 'uniform abrasion')
plt.plot(x, eta_var, label = 'variable abrasion')
plt.legend()
plt.show()
```
