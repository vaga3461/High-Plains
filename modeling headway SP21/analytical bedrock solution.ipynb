```python
# trying an analytical soln to bedrock model
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 0.01
c_param = 1
U = 0.002
x = np.arange(1, 100, 1)
q = x

# quadratic eqn variables
a = c_param * q
b = k * q
c = -U

S = (-b + (np.sqrt((b**2) - 4*a*c)))/(2*a)
```

```python
plt.plot(x, S)
plt.show()
```

```python
# the way shown above was wrong, because it used the following math:
```

$\frac{dz}{dt} = U - E$

$E = kqS + q_sS$, where $kqS$ is the plucking term and $q_sS$ should have been the abrasion term (and $q_s = cqS$). In this solution, if we assume $\frac{dz}{dt} = 0$ and solve for $S$, we get a solution in the form of a quadratic equation.

However, the abrasion term is not $q_sS$, but rather $\beta q_s$

Using the correct formulation for the abrasion term, we can re-write a steady state, analytical solution, meaning that we assume $\frac{dz}{dt} = 0$ and solve for S. This gives us the following profile.

```python
beta = 0.005

slope = U / ((k*q) + (beta*c_param*q))
```

```python
plt.plot(x, slope)
plt.show()
```

```python
# next up is an analytical solution for bedrock of two lithologies
# it looks the same, but now we implement arrays for k and beta instead of constant values
```

```python
k_array = np.zeros(len(x))
k_array[:50] = 0.001
k_array[50:] = 0.01

beta_array = np.zeros(len(x))
beta_array[:50] = 0.001
beta_array[50:] = 0.01

slope_two_lith = U / ((k_array*q) + (beta_array*c_param*q))
```

```python
plt.plot(x, slope_two_lith)
plt.show()
```
