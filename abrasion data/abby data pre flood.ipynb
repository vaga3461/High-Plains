```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('abby data csv.csv', header = 0)
```

```python
data
```

```python
fig, axs = plt.subplots(1, 2, figsize = (16, 4), dpi = 300)

#axs[0].scatter(data['Area'][0:6], data['Qb total'][0:6], marker = 'x', s = 60)
axs[0].scatter(data['Area'][0:6], data['Qb g'][0:6], marker = 'x', s = 60, label = 'granite')
axs[0].scatter(data['Area'][0:6], data['Qb lm'][0:6], marker = 'x', s = 60, label = 'light metamorphic')
axs[0].scatter(data['Area'][0:6], data['Qb dm'][0:6], marker = 'x', s = 60, label = 'dark metamorphic')
axs[0].scatter(data['Area'][0:6], data['Qb v'][0:6], marker = 'x', s = 60, label = 'volcanic')
axs[0].set_xscale('log')
axs[0].legend()

#axs[1].scatter(data['Area'][6:], data['Qb total'][6:], marker = 'x', s = 60)
axs[1].scatter(data['Area'][6:], data['Qb g'][6:], marker = 'x', s = 60, label = 'granite')
axs[1].scatter(data['Area'][6:], data['Qb lm'][6:], marker = 'x', s = 60, label = 'light metamorphic')
axs[1].scatter(data['Area'][6:], data['Qb dm'][6:], marker = 'x', s = 60, label = 'dark metamorphic')
axs[1].scatter(data['Area'][6:], data['Qb v'][6:], marker = 'x', s = 60, label = 'volcanic')
axs[1].scatter(data['Area'][6:], data['Qb ss'][6:], marker = 'x', s = 60, label = 'sandstone')
axs[1].set_xscale('log')
axs[1].legend()
```
